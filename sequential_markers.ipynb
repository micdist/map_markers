```python
from ipyleaflet import Map, Marker, Popup
import ipywidgets as widgets
from ipywidgets import HTML
from IPython.display import display
```

```python
coordinates_list = [
    (41.6874, 15.3797),
    (41.5083, 15.3343),
    (41.4622, 15.5446),
    (41.2664, 15.8940)
]
```

```python
# per centrare la mappa vicino alle coordinate
max_lat, max_lng = max(coordinates_list, key=lambda x: x[0])
min_lat, min_lng = min(coordinates_list, key=lambda x: x[0])
center_lat = (max_lat + min_lat)/2
center_lng = (max_lng + min_lng)/2
```

```python
center = (center_lat,center_lng)
m = Map(zoom=8,center=center)
# per centrare la mappa vicino alle coordinate
m.fit_bounds([(min_lat+0.10, min_lng+0.10), (max_lat+0.10, max_lng+0.10)])
display(m)


button = widgets.Button(description="Next Marker", button_style= 'success')

output = widgets.Output()
display(button, output)

marker_index = 0

def on_button_clicked(b):
    global marker_index
    if marker_index < len(coordinates_list):
        coordinates = coordinates_list[marker_index]
        posizione = str(marker_index+1)
        # title=posizione mostra l'ordine di aggiunta del marker se ci si ferma sopra col mouse
        # invisibile se  m.add_layer(popup)
        marker = Marker(location=coordinates,title='')
        m.add_layer(marker)
        marker_index += 1


        message1 = HTML()
        message1.value = f'{posizione}, {coordinates[0]} - {coordinates[1]}'



        popup =  Popup(
        location=coordinates,
        child=message1,
        min_width = 0,
        # max_width=5,
        max_height=29,
        close_button=True,
        auto_close=False,
        close_on_escape_key=False,

        show_arrow=True,
        )

        # visualizza ordine di aggiunta solo se click sul marker
        marker.popup = message1

        # visualizza il pop con l'ordine di aggiunta sopra il marker
        m.add_layer(popup)


        print("Marker added.")
    else:
       print("No more markers to add.")
       button.layout.visibility = "hidden"

button.on_click(on_button_clicked)
```
